```python
# load data
dataset = pd.read_csv('data02.csv')
dataset = np.array(dataset)
data_X = dataset[:,0:2]
data_y = dataset[:,2]
```

# Neumerical example
$$ y = 5 * x1 + 3 * x2 + error, error = N(0,1)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
# linear regression fitting 
linear_model = LinearRegression()
linear_model.fit(data_X, data_y)
lr_pred = linear_model.predict(data_X)

# true spec vs predicted spec
plt.scatter(data_y,lr_pred)
plt.plot(data_y,data_y,color ='r') 
plt.show()

mae = np.mean(np.abs(data_y-lr_pred))
r2 = r2_score(data_y,lr_pred)

print("mae = ", np.round(mae, 3))
print("r2 = ", np.round(r2,3) )
```

```python
# feature importances 

# Coefficient( Feature importances )
plt.figure(figsize=(6,4))
plt.bar(['X1','X2'],linear_model.coef_)
plt.title('Feature importances')       
plt.show()
linear_model.coef_
```

```python
data_X_cleaned = data_X[data_y < 50,:] 
data_y_cleaned = data_y[data_y < 50] 
```

```python
# linear regression fitting 
linear_model = LinearRegression()
linear_model.fit(data_X_cleaned, data_y_cleaned)
lr_pred = linear_model.predict(data_X_cleaned)

# true spec vs predicted spec
plt.scatter(data_y_cleaned,lr_pred)
plt.plot(data_y_cleaned,data_y_cleaned,color ='r')
plt.show()

mae = np.mean(np.abs(data_y_cleaned-lr_pred))
r2 = r2_score(data_y_cleaned,lr_pred)

print("mae = ", np.round(mae, 3))
print("r2 = ", np.round(r2,3) )
```

```python
# feature importances 

# Coefficient( Feature importances )
plt.figure(figsize=(6,4))
plt.bar(['X1','X2'],linear_model.coef_)
plt.title('Feature importances')       
plt.show()
linear_model.coef_
```
